```python
import numpy as np

# A complex number: a + bi
c1 = complex(3, 2)       # 3 + 2i
c2 = complex(1, 4)       # 1 + 4i

print("c1 =", c1)
print("c2 =", c2)

# Basic operations
print("\nAddition:", c1 + c2)
print("Multiplication:", c1 * c2)

# Real and imaginary parts
print("\nReal part of c1:", c1.real)
print("Imaginary part of c1:", c1.imag)

# Magnitude (distance from origin on the plane)
print("\nMagnitude of c1:", abs(c1))
```

```python
# Probability — simple but critical for quantum

outcomes = [0, 1]            # Like a qubit — either 0 or 1
probs    = [0.5, 0.5]        # Equal probability — superposition!

print("Outcomes:", outcomes)
print("Probabilities:", probs)
print("Sum of probabilities:", sum(probs))  # Must always equal 1

# Simulate 10 qubit measurements
measurements = np.random.choice(outcomes, size=10, p=probs)
print("\n10 qubit measurements:", measurements)
```

```python
measurements = np.random.choice(outcomes, size=1000, p=probs)
zeros = list(measurements).count(0)
ones  = list(measurements).count(1)
print(f"Zeros: {zeros}, Ones: {ones}")
```

```python
# Biased qubit — more likely to be 1
probs_biased = [0.2, 0.8]
measurements = np.random.choice(outcomes, size=20, p=probs_biased)
print("Biased measurements:", measurements)
```

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Dot product
dot = np.dot(a, b)
print("Dot product:", dot)

# Manual verification
manual = (1*4) + (2*5) + (3*6)
print("Manual calculation:", manual)
```

```python
# Same direction
a = np.array([1, 0])
b = np.array([1, 0])
print("Same direction:", np.dot(a, b))       # Expect: 1

# Opposite direction
c = np.array([1, 0])
d = np.array([-1, 0])
print("Opposite direction:", np.dot(c, d))   # Expect: -1

# Perpendicular
e = np.array([1, 0])
f = np.array([0, 1])
print("Perpendicular:", np.dot(e, f))        # Expect: 0
```

```python
# Cross product only works in 3D
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

cross = np.cross(a, b)
print("Cross product:", cross)
```
