```python
import numpy as np
import matplotlib.pyplot as plt

# Uniform distribution — like a fair dice
uniform = np.random.uniform(low=0, high=1, size=1000)

plt.hist(uniform, bins=20, color='steelblue', edgecolor='black')
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Uniform distribution — like a fair dice
uniform = np.random.uniform(low=0, high=1, size=1000)

plt.hist(uniform, bins=20, color='steelblue', edgecolor='black')
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Normal distribution — bell curve
normal = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(normal, bins=30, color='coral', edgecolor='black')
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, scale in zip(axes, [0.5, 1, 3]):
    data = np.random.normal(loc=0, scale=scale, size=1000)
    ax.hist(data, bins=30, color='coral', edgecolor='black')
    ax.set_title(f"Scale = {scale}")

plt.show()
```

```python
for scale, color in zip([0.5, 1, 3], ['blue', 'coral', 'green']):
    data = np.random.normal(loc=0, scale=scale, size=1000)
    plt.hist(data, bins=30, alpha=0.5, color=color, 
             edgecolor='black', label=f'Scale={scale}')

plt.title("Three Normal Distributions Compared")
plt.legend()
plt.show()
```
