```python
import math
import random

# Step 1 — Pick two prime numbers
p = 61
q = 53

# Step 2 — Calculate n (public modulus)
n = p * q
print(f"n = {p} x {q} = {n}")

# Step 3 — Calculate totient
totient = (p - 1) * (q - 1)
print(f"Totient = {totient}")

# Step 4 — Choose public exponent e
e = 17  # Must be coprime with totient
print(f"Public key: (e={e}, n={n})")

# Step 5 — Calculate private exponent d
d = pow(e, -1, totient)
print(f"Private key: (d={d}, n={n})")
```

```python
# Encrypt a number (RSA works on numbers)
message = 65  # Letter 'A' in ASCII

# Encryption — using public key (e, n)
ciphertext = pow(message, e, n)
print(f"Original message:  {message}")
print(f"Encrypted:         {ciphertext}")

# Decryption — using private key (d, n)
decrypted = pow(ciphertext, d, n)
print(f"Decrypted:         {decrypted}")

# Verify
print(f"Match: {message == decrypted}")
```

```python
# Encrypt a full message letter by letter
def rsa_encrypt_text(text, e, n):
    return [pow(ord(char), e, n) for char in text]

def rsa_decrypt_text(ciphertext, d, n):
    return ''.join([chr(pow(char, d, n)) for char in ciphertext])

message = "Hello"
encrypted = rsa_encrypt_text(message, e, n)
decrypted = rsa_decrypt_text(encrypted, d, n)

print(f"Original:  {message}")
print(f"Encrypted: {encrypted}")
print(f"Decrypted: {decrypted}")
```
