```python
import requests

# Your real IP
normal = requests.get("https://api.ipify.org?format=json")
print(f"Your real IP: {normal.json()['ip']}")

# Through Tor
proxies = {
    'http':  'socks5://127.0.0.1:9050',
    'https': 'socks5://127.0.0.1:9050'
}

tor = requests.get("https://api.ipify.org?format=json", proxies=proxies)
print(f"Your Tor IP:  {tor.json()['ip']}")
```

```python
# Geolocate the Tor exit node
tor_ip = "109.70.100.10"
location = requests.get(f"https://ipapi.co/{tor_ip}/json/")
data = location.json()

print(f"Tor Exit Node Location:")
print(f"IP:      {data['ip']}")
print(f"Country: {data['country_name']}")
print(f"City:    {data['city']}")
print(f"ISP:     {data['org']}")
```
