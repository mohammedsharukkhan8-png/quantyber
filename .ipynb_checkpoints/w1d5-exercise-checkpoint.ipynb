```python
import numpy as np

# Convert to radians first — Python uses radians not degrees
v = np.array([4 * np.cos(np.radians(310)), 4 * np.sin(np.radians(310))])
w = np.array([7 * np.cos(np.radians(50)),  7 * np.sin(np.radians(50))])

result = v + w
print("v =", np.round(v, 2))
print("w =", np.round(w, 2))
print("v + w =", np.round(result, 2))
```
