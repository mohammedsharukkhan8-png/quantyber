```python
import numpy as np

# --- DAY 1: Vectors & Matrix Multiplication ---

# Your first vector
v = np.array([1, 0])
print("My first vector:", v)

# A simple matrix
M = np.array([[1, 2],
              [3, 4]])
print("My first matrix:\n", M)

# Matrix × Vector
result = M @ v
print("Matrix applied to vector:", result)
```

```python
# --- DAY 2: Eigenvalues & Eigenvectors ---

# NumPy calculates eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(M)

print("Eigenvalues:", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)
```

```python
# --- PROOF: M @ v = λ × v ---

v0 = eigenvectors[:, 0]      # First eigenvector
λ0 = eigenvalues[0]          # First eigenvalue

left_side  = M @ v0
right_side = λ0 * v0

print("M @ v   =", left_side)
print("λ × v   =", right_side)
print("Are they equal?", np.allclose(left_side, right_side))
```
