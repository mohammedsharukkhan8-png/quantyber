```python
import subprocess

def dns_lookup(domain):
    print(f"\n{'='*50}")
    print(f"DNS Records for: {domain}")
    print('='*50)
    
    # A record (IPv4)
    result = subprocess.run(['dig', '+short', 'A', domain],
                          capture_output=True, text=True)
    print(f"A Record (IPv4):  {result.stdout.strip()}")
    
    # MX record (mail servers)
    result = subprocess.run(['dig', '+short', 'MX', domain],
                          capture_output=True, text=True)
    print(f"MX Record (Mail): {result.stdout.strip()}")
    
    # NS record (nameservers)
    result = subprocess.run(['dig', '+short', 'NS', domain],
                          capture_output=True, text=True)
    print(f"NS Record:        {result.stdout.strip()}")
    
    # TXT record
    result = subprocess.run(['dig', '+short', 'TXT', domain],
                          capture_output=True, text=True)
    print(f"TXT Record:       {result.stdout.strip()[:80]}...")

# Look up multiple domains
dns_lookup("google.com")
dns_lookup("github.com")
```

```python
import urllib.request
import urllib.error

def http_inspect(url):
    print(f"\n{'='*50}")
    print(f"HTTP Inspection: {url}")
    print('='*50)
    
    try:
        req = urllib.request.Request(url)
        req.add_header('User-Agent', 'Mozilla/5.0')
        
        with urllib.request.urlopen(req, timeout=5) as response:
            print(f"Status Code:     {response.status}")
            print(f"HTTP Version:    HTTP/1.1")
            print(f"\nResponse Headers:")
            for key, value in response.headers.items():
                print(f"  {key:30s}: {value}")
    except urllib.error.HTTPError as e:
        print(f"HTTP Error: {e.code} {e.reason}")
    except Exception as e:
        print(f"Error: {e}")

# Inspect real websites
http_inspect("https://github.com")
```
