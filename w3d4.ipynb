```python
import hashlib
import os

def hash_file(filepath):
    sha256 = hashlib.sha256()
    with open(filepath, 'rb') as f:
        # Read in chunks — handles large files
        for chunk in iter(lambda: f.read(4096), b""):
            sha256.update(chunk)
    return sha256.hexdigest()

# Create a test file
with open('test_file.txt', 'w') as f:
    f.write("This is Sharuk's secure document.")

# Hash it
original_hash = hash_file('test_file.txt')
print(f"Original hash:  {original_hash}")

# Simulate tampering
with open('test_file.txt', 'a') as f:
    f.write(" TAMPERED!")

# Hash again
tampered_hash = hash_file('test_file.txt')
print(f"Tampered hash:  {tampered_hash}")
print(f"\nFile integrity: {'✅ INTACT' if original_hash == tampered_hash else '❌ TAMPERED!'}")
```

```python
import socket

def scan_ports(host, port_range):
    print(f"Scanning {host} for open ports...")
    print(f"Port range: {port_range[0]} - {port_range[-1]}\n")
    
    open_ports = []
    
    for port in port_range:
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        sock.settimeout(0.5)
        
        result = sock.connect_ex((host, port))
        if result == 0:
            open_ports.append(port)
            print(f"Port {port:5d} → OPEN ✅")
        sock.close()
    
    print(f"\nTotal open ports found: {len(open_ports)}")
    return open_ports

# Scan localhost (your own machine — safe to scan)
scan_ports("127.0.0.1", range(1, 1025))
```

```python
from cryptography.fernet import Fernet

# Generate a secure key
key = Fernet.generate_key()
cipher = Fernet(key)

print(f"Encryption key: {key.decode()}")

# Encrypt a message
message = "Sharuk's secret: quantum security is the future"
encrypted = cipher.encrypt(message.encode())
print(f"\nOriginal:  {message}")
print(f"Encrypted: {encrypted.decode()}")

# Decrypt it back
decrypted = cipher.decrypt(encrypted).decode()
print(f"Decrypted: {decrypted}")

# Try decrypting with wrong key
wrong_key = Fernet.generate_key()
wrong_cipher = Fernet(wrong_key)
try:
    wrong_cipher.decrypt(encrypted)
except Exception as e:
    print(f"\nWrong key attempt: ❌ Decryption FAILED — {type(e).__name__}")
```

```python
# Encrypt the SAME message twice
msg = "Hello Sharuk"
enc1 = cipher.encrypt(msg.encode())
enc2 = cipher.encrypt(msg.encode())

print(f"Encryption 1: {enc1.decode()}")
print(f"Encryption 2: {enc2.decode()}")
print(f"Are they same? {enc1 == enc2}")
```
