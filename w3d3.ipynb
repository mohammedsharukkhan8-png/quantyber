```python
import ssl
import socket

def get_certificate_info(hostname):
    context = ssl.create_default_context()
    
    with socket.create_connection((hostname, 443)) as sock:
        with context.wrap_socket(sock, server_hostname=hostname) as ssock:
            cert = ssock.getpeercert()
    
    print(f"Website: {hostname}")
    print(f"\nIssued To:   {cert['subject'][0][0][1]}")
    print(f"Issued By:   {cert['issuer'][1][0][1]}")
    print(f"Valid From:  {cert['notBefore']}")
    print(f"Valid Until: {cert['notAfter']}")
    print(f"Version:     {cert['version']}")

# Check Google's certificate
get_certificate_info("google.com")
```

```python
# Check multiple websites
websites = ["github.com", "wikipedia.org", "anthropic.com"]

for site in websites:
    print(f"\n{'='*50}")
    get_certificate_info(site)
```
