```python
import socket

# Get your machine's hostname and IP
hostname = socket.gethostname()
ip_address = socket.gethostbyname(hostname)

print(f"Hostname:   {hostname}")
print(f"IP Address: {ip_address}")

# Resolve a domain name to IP (DNS lookup)
domains = ["google.com", "github.com", "anthropic.com"]

print(f"\nDNS Resolution:")
for domain in domains:
    ip = socket.gethostbyname(domain)
    print(f"{domain:20s} → {ip}")
```

```python
import socket
import time

def trace_connection(host, port):
    print(f"Connecting to {host}:{port}")
    
    start = time.time()
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    sock.settimeout(3)
    
    result = sock.connect_ex((host, port))
    end = time.time()
    
    if result == 0:
        print(f"✅ Connected successfully!")
        print(f"Response time: {(end-start)*1000:.2f}ms")
        
        # Get actual IP connected to
        ip = socket.gethostbyname(host)
        print(f"Server IP: {ip}")
    else:
        print(f"❌ Connection failed (code: {result})")
    
    sock.close()

# Test connections
print("="*40)
trace_connection("google.com", 443)    # HTTPS
print("="*40)
trace_connection("github.com", 443)    # HTTPS
print("="*40)
trace_connection("google.com", 80)     # HTTP
```
