```python
from scapy.all import *

# Capture 20 packets
print("Capturing 20 packets — browse something!")
packets = sniff(count=20, timeout=20)

print(f"\nCaptured {len(packets)} packets\n")

# Analyze each packet
print("="*60)
for i, pkt in enumerate(packets):
    if IP in pkt:
        src = pkt[IP].src
        dst = pkt[IP].dst
        proto = pkt[IP].proto
        proto_name = {6: 'TCP', 17: 'UDP', 1: 'ICMP'}.get(proto, str(proto))
        
        if TCP in pkt:
            sport = pkt[TCP].sport
            dport = pkt[TCP].dport
            flags = pkt[TCP].flags
            print(f"Packet {i+1:2d}: {src}:{sport} → {dst}:{dport} [{proto_name}] flags={flags}")
        elif UDP in pkt:
            sport = pkt[UDP].sport
            dport = pkt[UDP].dport
            print(f"Packet {i+1:2d}: {src}:{sport} → {dst}:{dport} [{proto_name}]")
```

```python
import socket
import requests

# Identify the server you were talking to
server_ip = "57.144.209.32"

# Reverse DNS lookup
try:
    hostname = socket.gethostbyaddr(server_ip)
    print(f"Server IP:       {server_ip}")
    print(f"Server hostname: {hostname[0]}")
except:
    print(f"No reverse DNS for {server_ip}")

# Geolocate it
location = requests.get(f"https://ipapi.co/{server_ip}/json/")
data = location.json()
print(f"Country: {data['country_name']}")
print(f"City:    {data['city']}")
print(f"ISP:     {data['org']}")
```
