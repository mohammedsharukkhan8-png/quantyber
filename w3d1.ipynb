```python
import hashlib

message = "Quantum is the future!"

# SHA-256 hash
sha256 = hashlib.sha256(message.encode()).hexdigest()
print(f"Message:      {message}")
print(f"SHA-256:      {sha256}")
print(f"Hash length:  {len(sha256)} characters")
```

```python
messages = [
    "Quantum is the future!",
    "quantum is the future!",  # lowercase q
    "Quantum is the future?",  # ? instead of !
    "Quantum is the future ",  # extra space
]

print("Avalanche Effect:\n")
for msg in messages:
    h = hashlib.sha256(msg.encode()).hexdigest()
    print(f"'{msg}'")
    print(f"→ {h}\n")
```

```python
import hashlib
import os

def hash_password(password):
    # Generate random salt
    salt = os.urandom(16).hex()
    # Hash password + salt
    hashed = hashlib.sha256((password + salt).encode()).hexdigest()
    return salt, hashed

def verify_password(password, salt, stored_hash):
    # Recreate hash with same salt
    hashed = hashlib.sha256((password + salt).encode()).hexdigest()
    return hashed == stored_hash

# Simulate registration
password = "Sharuk@123"
salt, stored_hash = hash_password(password)

print(f"Password:     {password}")
print(f"Salt:         {salt}")
print(f"Stored hash:  {stored_hash}")

# Simulate login
print(f"\nLogin attempt with correct password:")
print(f"Verified: {verify_password('Sharuk@123', salt, stored_hash)}")

print(f"\nLogin attempt with wrong password:")
print(f"Verified: {verify_password('sharuk@123', salt, stored_hash)}")
```
