```python
import subprocess
import time
import threading
import re
from collections import Counter
from datetime import datetime

def simulate_brute_force():
    """Simulate multiple failed SSH attempts"""
    print("🔴 SIMULATING: Brute Force Attack")
    print("Attempting fake SSH connections to localhost...\n")
    
    # Simulate failed login attempts in logs
    fake_ips = ["192.168.1.100", "10.0.0.50", "172.16.0.25"]
    usernames = ["root", "admin", "user", "test"]
    
    simulated_logs = []
    timestamp = datetime.now()
    
    for i in range(15):
        ip = fake_ips[i % len(fake_ips)]
        user = usernames[i % len(usernames)]
        log_line = f"Jul 20 {timestamp.strftime('%H:%M:%S')} server sshd[1234]: Failed password for {user} from {ip} port 22"
        simulated_logs.append(log_line)
        print(f"  ❌ Failed attempt {i+1}: {user}@{ip}")
        time.sleep(0.1)
    
    print(f"\n✅ Simulation complete — {len(simulated_logs)} attempts generated")
    return '\n'.join(simulated_logs)

# Run simulation
brute_force_logs = simulate_brute_force()
```

```python
def detect_brute_force(logs):
    """SIEM detection engine for brute force"""
    print(f"\n{'='*55}")
    print(f"🔍 SIEM: Analyzing Brute Force Simulation")
    print(f"{'='*55}")
    
    ip_counter = Counter()
    user_counter = Counter()
    attempts = []
    
    for line in logs.split('\n'):
        if "Failed password" in line:
            ip_match = re.search(r'from (\d+\.\d+\.\d+\.\d+)', line)
            user_match = re.search(r'for (\w+) from', line)
            if ip_match and user_match:
                ip = ip_match.group(1)
                user = user_match.group(1)
                ip_counter[ip] += 1
                user_counter[user] += 1
                attempts.append({"ip": ip, "user": user})
    
    # Analysis
    print(f"\n📊 ATTACK ANALYSIS:")
    print(f"Total attempts:    {len(attempts)}")
    print(f"Unique IPs:        {len(ip_counter)}")
    print(f"Targeted accounts: {list(user_counter.keys())}")
    
    print(f"\n🚨 THREAT DETECTION:")
    alerts_generated = []
    for ip, count in ip_counter.most_common():
        if count >= 3:
            alert = {
                "type":     "BRUTE_FORCE",
                "severity": "HIGH",
                "ip":       ip,
                "attempts": count,
                "action":   f"sudo ufw deny from {ip}"
            }
            alerts_generated.append(alert)
            print(f"  🔴 ALERT: {ip} → {count} attempts → BLOCK RECOMMENDED")
            print(f"     Action: {alert['action']}")
    
    print(f"\n💡 AUTOMATED RESPONSE:")
    print(f"  Alerts generated: {len(alerts_generated)}")
    print(f"  IPs to block:     {len(alerts_generated)}")
    print(f"  Response time:    {datetime.now().strftime('%H:%M:%S')}")
    
    return alerts_generated

# Run detection
alerts = detect_brute_force(brute_force_logs)
```

```python
def simulate_data_exfiltration():
    """Simulate suspicious outbound data transfer"""
    print(f"\n{'='*55}")
    print(f"🔴 SIMULATING: Data Exfiltration")
    print(f"{'='*55}")
    
    # Simulate large outbound connections
    suspicious_connections = [
        {"dst": "185.220.101.45", "port": 4444, "bytes": 5242880,  "process": "python3"},
        {"dst": "45.33.32.156",   "port": 1337, "bytes": 10485760, "process": "bash"},
        {"dst": "192.168.1.100",  "port": 8080, "bytes": 1048576,  "process": "curl"},
    ]
    
    print(f"\nDetecting suspicious outbound transfers:\n")
    
    alerts = []
    for conn in suspicious_connections:
        mb = conn['bytes'] / 1048576
        
        # Flag suspicious ports and large transfers
        suspicious_ports = [4444, 1337, 31337, 6666, 9999]
        is_suspicious_port = conn['port'] in suspicious_ports
        is_large_transfer  = conn['bytes'] > 1048576  # > 1MB
        
        if is_suspicious_port or is_large_transfer:
            alert = {
                "type":    "DATA_EXFILTRATION",
                "dst_ip":  conn['dst'],
                "port":    conn['port'],
                "size_mb": round(mb, 2),
                "process": conn['process']
            }
            alerts.append(alert)
            
            print(f"  🚨 SUSPICIOUS TRANSFER DETECTED!")
            print(f"     Process:     {conn['process']}")
            print(f"     Destination: {conn['dst']}:{conn['port']}")
            print(f"     Data size:   {mb:.1f} MB")
            
            if is_suspicious_port:
                print(f"     Reason:      ⚠️  Known C2 port ({conn['port']})")
            if is_large_transfer:
                print(f"     Reason:      ⚠️  Large outbound transfer")
            print()
    
    print(f"Total exfiltration alerts: {len(alerts)}")
    return alerts

exfil_alerts = simulate_data_exfiltration()
```
