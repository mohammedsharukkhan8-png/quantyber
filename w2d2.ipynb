```python
import numpy as np

# Simulate 10000 days
days = 10000

# 30% chance of rain each day
rain = np.random.choice([0, 1], size=days, p=[0.7, 0.3])

# 80% chance of cloudy if rainy, 20% chance of cloudy if not rainy
cloudy = np.where(rain == 1,
                  np.random.choice([0,1], size=days, p=[0.2, 0.8]),
                  np.random.choice([0,1], size=days, p=[0.8, 0.2]))
# P(rain) 
p_rain = np.mean(rain)
print(f"P(rain) = {p_rain:.2f}")

# P(cloudy)
p_cloudy = np.mean(cloudy)
print(f"P(cloudy) = {p_cloudy:.2f}")

# P(rain | cloudy) — days where BOTH rain and cloudy
p_rain_given_cloudy = np.mean(rain[cloudy == 1])
print(f"P(rain | cloudy) = {p_rain_given_cloudy:.2f}")
```

```python
# Verify using Bayes Theorem formula
# P(rain|cloudy) = P(cloudy|rain) × P(rain) / P(cloudy)

p_cloudy_given_rain = np.mean(cloudy[rain == 1])
print(f"P(cloudy|rain) = {p_cloudy_given_rain:.2f}")

# Bayes formula
bayes_result = (p_cloudy_given_rain * p_rain) / p_cloudy
print(f"\nBayes Theorem result:    {bayes_result:.2f}")
print(f"Direct calculation:      {p_rain_given_cloudy:.2f}")
print(f"Are they equal? {np.allclose(bayes_result, p_rain_given_cloudy, atol=0.01)}")
```

```python
import numpy as np

# ========================
# Bayes Theorem - Complete
# ========================

# Simulate 10000 days
days = 10000
np.random.seed(42)  # Makes results consistent every run

# 30% chance of rain each day
rain = np.random.choice([0, 1], size=days, p=[0.7, 0.3])

# 80% chance of cloudy if rainy, 20% chance of cloudy if not rainy
cloudy = np.where(rain == 1,
                  np.random.choice([0, 1], size=days, p=[0.2, 0.8]),
                  np.random.choice([0, 1], size=days, p=[0.8, 0.2]))

# Basic probabilities
p_rain               = np.mean(rain)
p_cloudy             = np.mean(cloudy)
p_rain_given_cloudy  = np.mean(rain[cloudy == 1])
p_cloudy_given_rain  = np.mean(cloudy[rain == 1])

print(f"P(rain)            = {p_rain:.2f}")
print(f"P(cloudy)          = {p_cloudy:.2f}")
print(f"P(rain|cloudy)     = {p_rain_given_cloudy:.2f}")
print(f"P(cloudy|rain)     = {p_cloudy_given_rain:.2f}")

# Bayes Theorem verification
bayes_result = (p_cloudy_given_rain * p_rain) / p_cloudy
print(f"\nBayes Theorem result:  {bayes_result:.2f}")
print(f"Direct calculation:    {p_rain_given_cloudy:.2f}")
print(f"Are they equal? {np.allclose(bayes_result, p_rain_given_cloudy, atol=0.01)}")
```

```python
def caesar_encrypt(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            result += chr((ord(char) - base + shift) % 26 + base)
        else:
            result += char
    return result

def caesar_decrypt(text, shift):
    return caesar_encrypt(text, -shift)

# Test it
message  = "Hello Sharuk"
shift    = 3

encrypted = caesar_encrypt(message, shift)
decrypted = caesar_decrypt(encrypted, shift)

print(f"Original:  {message}")
print(f"Encrypted: {encrypted}")
print(f"Decrypted: {decrypted}")
```

```python
def caesar_encrypt(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            result += chr((ord(char) - base + shift) % 26 + base)
        else:
            result += char
    return result

def caesar_decrypt(text, shift):
    return caesar_encrypt(text, -shift)

# Test it
message  = "Hello titan"
shift    = 3

encrypted = caesar_encrypt(message, shift)
decrypted = caesar_decrypt(encrypted, shift)

print(f"Original:  {message}")
print(f"Encrypted: {encrypted}")
print(f"Decrypted: {decrypted}")
```

```python
# What happens when shift wraps around Z?
print(caesar_encrypt("XYZ", 3))   # Should wrap around to ABC
print(caesar_encrypt("xyz", 3))   # Lowercase too

# What if shift is 26? (full circle)
print(caesar_encrypt("Hello", 26))

# Negative shift
print(caesar_encrypt("Hello", -3))
```

```python
# Brute force — try all 26 shifts
ciphertext = "Khoor Titan"

print("Brute force attack:")
for shift in range(26):
    attempt = caesar_decrypt(ciphertext, shift)
    print(f"Shift {shift:2d}: {attempt}")
```
